# 整合

```
from IPython.display import display, Markdown
import ipywidgets as widgets
import requests
import json

# Define global variables
url = "https://www.runoob.com/try/compile2.php"
headers = {
    "Content-Type": "application/x-www-form-urlencoded"
}
token = "066417defb80d038228de76ec581a50a"
languages = {
    "Go": {"code": 'package main\n\nimport "fmt"\n\nfunc main() {\n   fmt.Println("Hello, World")\n}', "language": "6", "fileext": "go"},
    "Python": {"code": 'print("Hello, World")', "language": "4", "fileext": "py"},
    "C": {"code": '#include <stdio.h>\n\nint main()\n{\n   /*  Write C code in this online editor and run it. */\n   printf("Hello, world! \\n");\n   return 0;\n}',"language": "7","fileext": "c"}
    # Add more languages here...
}

# Define a Markdown widget for explanatory text
explanation_text = widgets.Output(layout=widgets.Layout(width="50%"))
with explanation_text:
    display(Markdown('### Code Editor\n\nPlease select a language, then type your code in the text area below. \
                     Click the "Run" button to execute your code. \
                     If you want to reset the code to the default, please click the "Reset" button.'))

# Define a dropdown for language selection
language_dropdown = widgets.Dropdown(
    options=languages.keys(),
    value=list(languages.keys())[0],
    description='Language:',
)

# Define text areas for input and output
code_input = widgets.Textarea(
    value=languages[language_dropdown.value]["code"],
    placeholder='Type your code here',
    rows=10,
    layout=widgets.Layout(width="50%")
)

output_area = widgets.Output(layout=widgets.Layout(width="50%", border='1px solid black', padding='10px'))

# Define buttons to trigger the request and reset the input
run_button = widgets.Button(
    description='Run',
    button_style='success',
)

reset_button = widgets.Button(
    description='Reset',
    button_style='warning',
)

def run_code(b):
    data = {
        "code": code_input.value,
        "token": token,
        "stdin": "",
        "language": languages[language_dropdown.value]["language"],
        "fileext": languages[language_dropdown.value]["fileext"]
    }
    response = requests.post(url, headers=headers, data=data)
    response_dict = json.loads(response.text)

    # Clear the previous output and print the new one
    output_area.clear_output()
    with output_area:
        if response_dict["output"]:
            print("Output: ", response_dict["output"].strip())
        if response_dict["errors"].strip():
            print("Errors: ", response_dict["errors"].strip())

def reset_code(b):
    code_input.value = languages[language_dropdown.value]["code"]

run_button.on_click(run_code)
reset_button.on_click(reset_code)

# Define a function to update the default code when the selected language changes
def update_default_code(change):
    code_input.value = languages[change['new']]["code"]

# Attach the function as a handler to the 'value' property of the dropdown
language_dropdown.observe(update_default_code, names='value')

# Create a VBox to hold the widgets
box = widgets.VBox([explanation_text, language_dropdown, code_input, widgets.HBox([run_button, reset_button]), output_area])

# Display everything
display(box)
```
